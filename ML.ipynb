```python
# Función lambda para asignar valores de sentimiento
def assign_sentiment(stars):
    if stars == 3.0:
        return "neutral"
    elif stars > 3.0:
        return "positivo"
    else:
        return "negativo"

# Agregar nueva columna "sentiment" utilizando apply()
businessml["sentiment"] = businessml["stars"].apply(lambda x: assign_sentiment(x))
```

```python
import pandas as pd


def get_best_categories_by_state(state):
    # Filtrar el dataframe por el estado dado
    state_df = merged_df[merged_df['state'] == state]

    # Calcular el promedio de sentimientos por categoría en cada ciudad
    state_df['sentiment_review'] = state_df['sentiment_review'].map({'positive': 1, 'neutral': 0, 'negative': -1})
    city_category_sentiment = state_df.groupby(['state', 'city', 'categories'])['sentiment_review'].mean().reset_index()

    # Ordenar las categorías por el promedio de sentimientos descendente en cada ciudad
    city_category_sentiment_sorted = city_category_sentiment.sort_values(['state', 'city', 'sentiment_review'], ascending=[True, True, False])
    city_category_sentiment_sorted.drop_duplicates(subset=['state', 'city'], keep='first', inplace=True)

    # Obtener las categorías con mejores reseñas en cada ciudad
    best_categories_by_state = city_category_sentiment_sorted[['city', 'categories']]

    return best_categories_by_state

# Ejemplo de uso de la función
state = 'Hawaii'
best_categories = get_best_categories_by_state(state)

print("Mejores categorías en el estado", state, ":")
print(best_categories)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier

# Preprocesamiento de texto: convertir las reseñas en características numéricas
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(merged_df['review'])

# Definir las características y la variable objetivo
features = X
target = merged_df['categories']

# Entrenar el clasificador RandomForest
clf = RandomForestClassifier()
clf.fit(features, target)

```

```python
# Función de recomendación
def recommend_restaurants(state):
    # Obtener las mejores categorías para abrir un restaurante en el estado dado
    state_reviews = merged_df[merged_df['state'] == state]['review']
    state_features = vectorizer.transform(state_reviews)
    predicted_categories = clf.predict(state_features)
    unique_categories, category_counts = np.unique(predicted_categories, return_counts=True)
    sorted_categories = [x for _, x in sorted(zip(category_counts, unique_categories), reverse=True)]

    # Obtener las 5 primeras categorías recomendadas
    recommended_categories = sorted_categories[:5]

    # Obtener las ciudades donde esas categorías son populares
    recommended_cities = []
    for category in recommended_categories:
        cities = merged_df[(merged_df['state'] == state) & (merged_df['categories'] == category)]['city'].unique()
        recommended_cities.extend(cities[:3])  # Obtener las 3 primeras ciudades para cada categoría recomendada

    return recommended_categories, recommended_cities

# Ejemplo de uso
state = 'California'
recommended_categories, recommended_cities = recommend_restaurants(state)

print(f"Mejores 5 categorías para abrir un restaurante en {state}:")
for category in recommended_categories:
    print(category)

print("\nCiudades donde esas categorías son populares:")
for city in recommended_cities:
    print(city)
```
